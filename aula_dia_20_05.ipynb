```python
#Você foi contratado para desenvolver um sistema simples de caixa eletrônico

# O sistema deverá funcionar através de um menu interativo utilizando a estrutura de repetição while.

# O programa deve iniciar com um saldo inicial de R$ 1.000,00 e permitir que o usuário escolha diferentes operações até decidir sair do sistema

# Funcionalidades:

# 1 - ver saldo;
# 2 - depositar dinheiro;
# 3 - sacar dinheiro;
# 4 - sair

# Regras

# ver saldo
# ao escolher essa opção, o sistema deverá mostrar o saldo atual da conta;

# Depositar dinheiro
# o usuário deverá informar um valor para depósito
# o valor digitado deve ser somado ao saldo

# Sacar dinheiro
# o usuário deverá informar o valor do saque

# O sistema deverá verificar:
# se houver saldo suficiente - realizar o saque
# se não houver saldo suficiente - mostrar mensagem "saldo insuficiente"

# Sair

# o programa deverá encerrar apenas quando o usuário escolher a opção


saldo = 1000.00
opcao = 0

while opcao != 4:
    print("--- CAIXA ELETRÔNICO ---")
    print("1 - Ver saldo")
    print("2 - Depositar dinheiro")
    print("3 - Sacar dinheiro")
    print("4 - Sair")

    opcao = int(input("Escolha uma opção: "))

    if opcao == 1:
        print(f"Seu saldo atual é: R$ {saldo:.2f}")

    elif opcao == 2:
        valor_deposito = float(input("Digite o valor do depósito: R$ "))
        if valor_deposito > 0:
            saldo += valor_deposito
            print("Depósito realizado com sucesso!")
        else:
            print("Valor inválido para depósito.")

    elif opcao == 3:
        valor_saque = float(input("Digite o valor do saque: R$ "))
        if 0 < valor_saque <= saldo:
            saldo -= valor_saque
            print("Saque realizado com sucesso!")
        else:
            print("Saldo insuficiente ou valor inválido.")

    elif opcao == 4:
        print("Encerrando o sistema. Até logo!")

    else:
        print("Opção inválida. Tente novamente.")

```
