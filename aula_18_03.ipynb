```python
# Operações aritiméticas

#Fazer um programa que o usuário digita 02 números
#Saída mostrar operações aritiméticas

#Entrada

a = int(input("digite um número :"))
b = int(input("digite um número :"))


# Processamento

soma = a + b
subtração = a - b
multiplicacao = a * b
divisao = a / b
div_int = a // b # Alterado para divisão inteira correta
reto = a % b
potencia = a ** b

# Saida

print("")
print("------------------CALCULADORA-----------------------")
print("                                                                ")
print("--------------------------SOMA----------------------------------")
print(f"o resultado da SOMA de {a} MAIS {b} é igual a : {soma}")
print("                                                                ")
print("--------------------------SUBTRAÇÃO-----------------------------")
print(f"o resultado da SUBTRAÇÃO de {a} MENOS {b} é igual a : {subtração}")
print("                                                                ")
print("-----------------------MULTIPLICAÇÃO------------------------------")
print(f"o resultado da SOMA de {a} VEZES  {b} é igual a : {multiplicacao}")
print("                                                                ")
print("--------------------------DIVISÃO-----------------------------")
print(f"o resultado da DIVISÃO de {a} DIVIDIDO {b} é igual a : {divisao}") # Corrigido aqui
print("                                                                ")
print("--------------------------DIVISÃO INTEIRA-------------------------")
print(f"o resultado da DIVISÃO INTEIRA de {a} DIVIDIDO {b} é igual a : {div_int}")
print("                                                                ")
print("--------------------------RESTO-------------------------")
print(f"o resultado da RESTO de {a} DIVIDIDO {b} é igual a : {reto}") # Corrigido aqui
print("                                                                ")
print("--------------------------POTENCIA-------------------------")
print(f"o resultado da POTENCIA de {a} ELEVADO {b} é igual a : {potencia}")
```
