```python
#Fazer um programa que converte temperatura

#Entrada

celcius = int(input("digite a temperatura em celsius"))

#Processamento

F = (celcius * 9/5) + 32

#Saida

print(f"A temperatura em Fahrenheit é {F}° graus")

```

```python
# Operações aritiméticas

#Fazer um programa que o usuário digita 02 números
#Saída mostrar operações aritiméticas

#Entrada

a = int(input("digite um número :"))
b = int(input("digite um número :"))


# Processamento

soma = a + b
subtração = a - b
multiplicacao = a * b
divisao = a / b
div_int = a // b
reto = a % b
potencia = a ** b

# Saida

print("")
print("------------------CALCULADORA-----------------------")
print("                                                                ")
print("--------------------------SOMA----------------------------------")
print(f"o resultado da SOMA de {a} MAIS {b} é igual a : {soma}")
print("                                                                ")
print("--------------------------SUBTRAÇÃO-----------------------------")
print(f"o resultado da SUBTRAÇÃO de {a} MENOS {b} é igual a : {subtração}")
print("                                                                ")
print("-----------------------MULTIPLICAÇÃO------------------------------")
print(f"o resultado da SOMA de {a} VEZES  {b} é igual a : {multiplicacao}")
print("                                                                ")
print("--------------------------DIVISÃO-----------------------------")
print(f"o resultado da DIVISÃO de {a} DIVIDIDO {b} é igual a : {divisao}")
print("                                                                ")
print("--------------------------DIVISÃO INTEIRA-------------------------")
print(f"o resultado da DIVISÃO INTEIRA de {a} DIVIDIDO {b} é igual a : {div_int}")
print("                                                                ")
print("--------------------------RESTO-------------------------")
print(f"o resultado da RESTO de {a} DIVIDIDO {b} é igual a : {reto}")
print("                                                                ")
print("--------------------------POTENCIA-------------------------")
print(f"o resultado da POTENCIA de {a} ELEVADO {b} é igual a : {potencia}")
```

```python
#Fazer um programa que converte horas em minutos e horas em segundos

#entrada

hora = int(input("digite a hora: "))

#processamento

minutos = hora * 60
segundos = hora * 3600

#saida

print(f"{hora} horas é igual a {minutos} minutos")
print(f"{hora} horas é igual a {segundos} segundos")
```

```python
# Fazer um programa que converte km em metros

#entrada

km = int(input("digite a distancia: "))

#processamento

metros = km * 1000

#saida

print(f"{km} km é igual a {metros} metros")
```

```python
#Fazer um programa que converte km em milhas

#entrada
km = float(input("digite a distancia: "))

#processamento

milhas = km / 1,609

#saida

print(f"{km} km é igual a {milhas} milhas")
```
