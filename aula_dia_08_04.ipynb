```python
# Fazer um código que identifique se o número é par ou impar

# Entrada
num = int(input("digite um número:"))

# Processamento
if num % 2 == 0:
    print(f"O número {num} é par")
else:
    print(f"{num} é impar")
```

```python
# Escrever um programa que o usuário digita seu nome e as suas 04 notas referente as bimestres,calcula a média e mostra se o aluno foi:
# aprovado, recuperação ou reprovado

# nota maior ou igual a 7 aprovado
# nota maior ou igual a 5 e menor que 7 recuperação
# nota menor que 5 reprovado

# Entrada
nome = str(input("digite seu nome:"))


nota1 = float(input("digite sua nota do 1º bimestre:"))
if nota1 > 10:
  print("digite uma nota válida (notas de 0 a 10)")

nota2 = float(input("digite sua nota do 2º bimestre:"))
if nota2 > 10:
  print("digite uma nota válida (notas de 0 a 10)")

nota3 = float(input("digite sua nota do 3º bimestre:"))
if nota3 > 10:
  print("digite uma nota válida (notas de 0 a 10)")

nota4 = float(input("digite sua nota do 4º bimestre:"))
if nota4 > 10:
  print("digite uma nota válida (notas de 0 a 10)")

média = (nota1 + nota2 + nota3 + nota4) / 4


# Processamento
if nota1 > 10 or nota2 > 10 or nota3 > 10 or nota4 > 10:
  print("digite uma nota válida (notas de 0 a 10)")

elif média >= 7:
  print(f"{nome} sua média {média} você foi aprovado")

elif média >= 5:
  print(f"{nome} sua média {média} você está de recuperação")

else:
  print(f"{nome} sua média {média} você foi reprovado")


```
