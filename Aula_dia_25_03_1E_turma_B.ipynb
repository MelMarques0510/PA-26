```python
# Fazer uma calculadora simples com 04 operações

print("----------MENU---------")
print("")
print("opção - 1 = SOMA")
print("opção - 2 = SUBTRAÇÃO ")
print("opção - 3 = MULTIPLICAÇÃO")
print("opção - 4 = DIVISÃO")

opcao = int(input("digite sua opção"))

# Entrada

a = float (input("digite um numero: "))
b = float (input("digite um numero: "))

# Processamento

if opcao == 1:
  print(f"a soma é igual a {a + b}")
  print("")

elif opcao == 2:
  print(f"a subtração é igual a {a - b}")
  print("")

elif opcao == 3:
  print(f"a multiplicação é igual a {a * b}")
  print("")

elif opcao == 4:
    if b != 0:
        print(f"a divisão é igual a {a / b}")
        print("")
    else:
        print("Erro: Divisão por zero não é permitida!")
        print("")

else:
    print("Opção inválida!")
    print("")

  # Saida
```
