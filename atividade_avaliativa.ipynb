```python
#PAR OU IMPAR

#Entrada

num = int(input("digite um número"))

#Processamento

if num % 2 == 0:
  print(f"o número {num} é par")
else:
  print(f"o número {num} é impar")
```

```python
#MAIOR DE DOIS NÚMEROS

#Entrada

a = int(input("digite um número: "))
b = int(input("digite um número: "))

#Processamento

if a > b:
  print(f"o número {a} é maior que o número {b}")
elif b > a:
  print(f"o número {b} é maior que o número {a}")
else:
  print(f"os dois números são iguais")
```

```python
#APROVAÇÃO DE ALUNO

#Entrada

nota1 = float(input("digite a nota do 1º bimestre: "))
nota2 = float(input("digite a nota do 2º bimestre: "))
nota3 = float(input("digite a nota do 3º bimestre: "))
nota4 = float(input("digite a nota do 4º bimestre: "))

#Processamento

resultado = (nota1 + nota2 + nota3 + nota4) / 4

if resultado >= 7:
  print(f"sua média de notas foi {resultado} você foi aprovado")
elif resultado >= 5:
  print(f"sua média de notas foi {resultado} você está de recuperação")
else:
  print(f"sua média de notas foi {resultado} você foi reprovado")
```

```python
#CLASSIFICAÇÃO DE IDADE

print("criança (0-12)")
print("adolescente (13-17)")
print("adulto (18-59)")
print("idoso (60+)")

#Entrada
a = int(input("digite sua idade: "))

#Processamento

if a <= 12:
  print(f"você tem {a} anos, você é criança")
elif a > 12 and a <= 17:
  print(f"você tem {a} anos, você é adolescente")
elif a > 17 and a <= 59:
  print(f"você tem {a} anos, você é adulto")
else:
  print(f"você tem {a} anos, você é idoso")

```

```python
#POSITIVO, NEGATIVO OU ZERO

#Entrada
a = int(input("digite um número: "))

#Processamento
if a > 0:
  print(f"seu número é positivo")
elif a < 0:
  print(f"seu número é negativo")
else:
  print(f"seu número é zero")
```

```python
#CALCULADORA SIMPLES

print("------OPERAÇÕES------")
print("")
print("opção 1 - soma")
print("opção 2 - subtração")
print("opção 3 - multiplicação")
print("opção 4 - divisão")

opção = int(input("digite sua opção: "))

#Entrada

a = float(input("digite seu número: "))
b = float(input("digite seu número: "))

#Processamento

if opção == 1:
  print(f"a soma é igual a {a + b}")
elif opção == 2:
  print(f"a subtração é igual a {a - b}")
elif opção == 3:
  print(f"a multiplicação é igual a {a * b}")
elif opção == 4:
  if b != 0:
    print(f"a divisão é igual a {a / b}")
  else:
    print("Erro: divisão por zero não é permitida: ")
else:
  print(f"opção inválida")
```

```python
#DIA DA SEMANA

#Entrada
d = int(input("digite um número de 1 a 7: "))

#Processamento

if d == 1:
  print("hoje é domingo")
elif d == 2:
  print(f"hoje é segunda")
elif d == 3:
  print(f"hoje é terça")
elif d == 4:
  print(f"hoje é quarta")
elif d == 5:
  print(f"hoje é quinta")
elif d == 6:
  print(f"hoje é sexta")
else:
  print(f"hoje é sabado")
```

```python
#ANO BISSEXTO

#Entrada
ano = int(input("digite o ano: "))

#Processamento
if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
  print(f"o ano {ano} é um ano bissexto")
else:
  print(f"o ano {ano} não é um ano bissexto ")
```

```python
#DESCONTO EM COMPRAS

# Entrada
valor_da_compra = float(input("Digite o valor da compra por favor: "))

# Processamento

if valor_da_compra >= 100:
  desconto = 0.10 # 10%
elif valor_da_compra >= 50:
  desconto = 0.05 # 5%
else:
  desconto = 0 # Sem desconto

valor_final = valor_da_compra * (1 - desconto)
print(f"O valor final da compra com desconto é: R${valor_final:.2f}")
```
