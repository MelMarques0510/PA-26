```python
# 1 - contagem simples, o usuário digita um número e o programa imprime do zero até o número

#entrada

n = int(input("digite um número: "))

#processamento

for i in range(n):
  print(i + 1)
```

```python
#2 - contagem regressiva, o usuário digita um número e o programa imprime do número até o zero

#entrada

n = int(input("digite um número: "))

#processamento

for i in range(n,-1,-1):
  print(i)
```

```python
# 3 - números pares, o usuário digita um número e o programa imprimi os números pares

#entrada

n = int(input("digite um número: "))

#processamento

for i in range(n + 1):
  if(i % 2 == 0):
    print(i)
```

```python
# 4 - números impares, o usuário digita um número e o programa imprimi os números impares

#entrada

n = int(input("digite um número: "))

#processamento

for i in range(n + 1):
  if i % 2 != 0:
    print(i)
```

```python
# 5 - soma, o usúario digita um número e o programa soma os números do zero até o número

#entrada

n = int(input("digite um número: "))
soma = 0

#processamento

for i in range(1, n + 1):
  soma += i
  print(f"{soma}")
```

```python
# 6 - tabuada, o usuário digita um número e o programa imprime a tabuada do número

#entrada

n = int(input("digite um número: "))

#processamento

for i in range(n + 1):
  print(f"{n} x {i} = {n * i}")
```

```python
# 8 - contar vogais, o usuário digita uma palavra e o programa conta as vogais

#entrada

p = str(input("digite uma palavra: "))
vogais = "aeiou"
letra = 0

for i in range (len(p)):
  if p[i] in vogais:
    letra += 1

print(f"A palavra tem {letra} vogais.")
```

```python
# 9 - maior número digitado, o usuário digita quatro números e o programa imprime o maior número digitado

#entrada

a = int(input("digite um número: "))
b = int(input("digite um número: "))
c = int(input("digite um número: "))
d = int(input("digite um número: "))

#processamento
numeros = [a, b, c, d]
maior_numero = max(numeros)

print(f"O maior número digitado é: {maior_numero}")
```
